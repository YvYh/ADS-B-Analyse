```python
from opensky_api import OpenSkyApi
import time as t
import pandas as pd
```

```python
time = 1536054850
api = OpenSkyApi()
ZSPD = (26.0836, 36.0836, 116.4819, 126.4819)
l=[]
states = api.get_states(bbox=ZSPD)
for s in states.states:
    #print("\t%r\t%r\t%r\t%r\t%r" %(s.callsign, s.longitude, s.latitude, s.geo_altitude, s.heading))
    l.append([s.callsign, s.longitude, s.latitude, s.geo_altitude, s.heading])
col=['callsign', 'lon','lat','geoaltitude','heading']
data=pd.DataFrame(l,columns=col)
data
```

```python
api = OpenSkyApi()
ZSPD = (26.0836, 36.0836, 116.4819, 126.4819)
l=[]
for i in range(3):
    states = api.get_states(bbox=ZSPD)
    print("Time{}\n".format(states.time))
    for s in states.states:
        l.append([s.callsign, s.longitude, s.latitude, s.geo_altitude, s.heading, s.icao24])
    t.sleep(30)
col=['callsign', 'lon','lat','geoaltitude','heading', 'icao24']
data=pd.DataFrame(l,columns=col)
data = data.drop_duplicates(subset=['lat','lon'])
data
```
